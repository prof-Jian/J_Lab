```python
unlabeled_1_ls = 'E:\\github lab\\data\\ren_task\\unlabeled-1\\病史特点\\病史特点-'
unlabeled_2_ls = 'E:\\github lab\\data\\ren_task\\unlabeled-2\\病史特点\\病史特点-'
t_dataset_ls = 'E:\\github lab\\ren_task\\training dataset v4\\02-病史特点\\'
'''
从上面我们看出，需要对raw做这么几件事：
- 去掉空格：txt.strip()
- 将‘、’替换成‘.'，我觉得将数字后面的’‘.’换成‘、’更准确:str.replace("1.", "1、")、str.replace("2.", "2、")
'''
def txt_clean(string):
    seq = string.replace(" ", "")
    seq = seq.replace("1.", "1、")
    seq = seq.replace("2.", "2、")
    seq = seq.replace("3.", "3、")
    seq = seq.replace("①", "1、")
    seq = seq.replace("2.", "2、")
    seq = seq.replace('style=','')
    seq = seq.replace('font-size:18px;','')
    seq = seq.replace('font-family:','')
    seq = seq.replace('&lt;','')
    seq = seq.replace('p&gt;','')
    seq = seq.replace('&ldquo;','')
    seq = seq.replace('&amp;nbsp;','')
    seq = seq.replace('span','')
    seq = seq.replace('&gt;','')
    seq = seq.replace('/span&gt;','')
    seq = seq.replace('style=&quot;','')
    seq = seq.replace('text-align:','')
    seq = seq.replace('center;','')
    seq = seq.replace('&quot;','')
    seq = seq.replace('&gt;','')
    seq = seq.replace('&rdquo','')
    seq = seq.replace('楷体_GB2312','')
    seq = seq.replace('EN-US&','')
    seq = seq.replace('color:','')
    seq = seq.replace('rgb(85,85,85)','')
    seq = seq.replace('color:rgb(0,0,0)','')
    seq = seq.replace('font-size:12px;','')
    
    if seq.startswith('、'):
        seq = seq[1:]
    seq = seq.strip()
    if seq == '':
        return None
    else:
        return seq
    
# 这里要是有个强力正则表达就好了
# 检查字符串是否包含中文
def isContainChinese(s):
    for c in s:
        if ('\u4e00' <= c <= '\u9fa5'):
            return True
    return False

from os.path import join
# 这里有一个缺陷，没检查文件里会不会是“nan”
unlabeled_1_text_list = []
for i in range(401,801):
    with open(unlabeled_1_ls+str(i), 'r', encoding='utf-8') as f:
        for line in f:
            a = txt_clean(line)
            if a == None or (not isContainChinese(a[:4])):
                continue
            else:
                unlabeled_1_text_list.append(a)
            

# 不知为何这个list最后只有2110个
unlabeled_2_text_list = []
for i in range(1,2206):
    with open(unlabeled_2_ls+str(i), 'r', encoding='utf-8') as f:
        for line in f:
            a = txt_clean(line)
            if a == None or (not isContainChinese(a[:4])):
                continue
            else:
                unlabeled_2_text_list.append(a)

t_dataset_text_list = []
for i in range(300):
    with open(join(t_dataset_ls,'病史特点-'+str(i+1)+'.txtoriginal.txt'), 'r', encoding='utf-8') as f:
        for line in f:
            a = txt_clean(line)
            if a == None or (not isContainChinese(a[:4])):
                continue
            else:
                t_dataset_text_list.append(a)
```

## 制作vocabulary和Vectors

```python
class Vocabulary(object):
    def __init__(self, token_to_idx=None):       
        if token_to_idx is None:
            token_to_idx = {}
        self._token_to_idx = token_to_idx

        self._idx_to_token = {idx: token 
                              for token, idx in self._token_to_idx.items()}
           
    def to_serializable(self):
        """ returns a dictionary that can be serialized """ 
        return {'token_to_idx': self._token_to_idx}

    @classmethod
    def from_serializable(cls, contents):
        """ instantiates the Vocabulary from a serialized dictionary """
        return cls(**contents) 

    def add_token(self, token):
        """Update mapping dicts based on the token.

        Args:
            token (str): the item to add into the Vocabulary
        Returns:
            index (int): the integer corresponding to the token
        """
        if token in self._token_to_idx:
            index = self._token_to_idx[token]
        else:
            index = len(self._token_to_idx)
            self._token_to_idx[token] = index
            self._idx_to_token[index] = token
        return index
            
    def add_many(self, tokens):
        """Add a list of tokens into the Vocabulary
        
        Args:
            tokens (list): a list of string tokens
        Returns:
            indices (list): a list of indices corresponding to the tokens
        """
        return [self.add_token(token) for token in tokens]

    def lookup_token(self, token):
        """Retrieve the index associated with the token 
        
        Args:
            token (str): the token to look up 
        Returns:
            index (int): the index corresponding to the token
        """
        return self._token_to_idx[token]

    def lookup_index(self, index):
        """Return the token associated with the index
        
        Args: 
            index (int): the index to look up
        Returns:
            token (str): the token corresponding to the index
        Raises:
            KeyError: if the index is not in the Vocabulary
        """
        if index not in self._idx_to_token:
            raise KeyError("the index (%d) is not in the Vocabulary" % index)
        return self._idx_to_token[index]

    def __str__(self):
        return "<Vocabulary(size=%d)>" % len(self)

    def __len__(self):
        return len(self._token_to_idx)
```

```python
class SequenceVocabulary(Vocabulary):
    def __init__(self, token_to_idx=None, unk_token="<UNK>",
                 mask_token="<MASK>",
                 end_seq_token="<END>"):

        super(SequenceVocabulary, self).__init__(token_to_idx)

        self._mask_token = mask_token
        self._unk_token = unk_token
        self._end_seq_token = end_seq_token

        self.mask_index = self.add_token(self._mask_token)
        self.unk_index = self.add_token(self._unk_token)
        self.end_seq_index = self.add_token(self._end_seq_token)

    def to_serializable(self):
        contents = super(SequenceVocabulary, self).to_serializable()
        contents.update({'unk_token': self._unk_token,
                         'mask_token': self._mask_token,
                         'end_seq_token': self._end_seq_token})
        return contents

    def lookup_token(self, token):
        """Retrieve the index associated with the token 
          or the UNK index if token isn't present.
        
        Args:
            token (str): the token to look up 
        Returns:
            index (int): the index corresponding to the token
        Notes:
            `unk_index` needs to be >=0 (having been added into the Vocabulary) 
              for the UNK functionality 
        """
        if self.unk_index >= 0:
            return self._token_to_idx.get(token, self.unk_index)
        else:
            return self._token_to_idx[token]
```

```python
import jieba
jieba.load_userdict("tag_list.txt")
jieba.add_word('1、')
jieba.add_word('2、')
jieba.add_word('3、')
jieba.add_word('4、')
jieba.add_word('5、')
jieba.add_word('mmHg')
jieba.add_word('&deg')
jieba.add_word('承德市')
jieba.add_word('双滦区')
jieba.add_word('CP')
jieba.add_word('既往')
```

```python
class Vectorizer(object):
    """ The Vectorizer which coordinates the Vocabularies and puts them to use"""   
    def __init__(self, char_vocab):
        """
        Args:
            char_vocab (Vocabulary): maps characters to integers
        """
        self.char_vocab = char_vocab

    def vectorize(self, surname, vector_length= Seq_Length):
        """
        Args:
            title (str): the string of characters
            vector_length (int): an argument for forcing the length of index vector
        """
        indices = []  # indices = [self.char_vocab.begin_seq_index]
        indices.extend(self.char_vocab.lookup_token(token) 
                       for token in jieba.cut(surname))
        indices.append(self.char_vocab.end_seq_index)

        if vector_length < 0:
            vector_length = len(indices)

        out_vector = np.zeros(vector_length, dtype=np.int64)# 没有规定长度的话，是多长就输出多长的列表；
        if len(indices) < vector_length:
            out_vector[:len(indices)] = indices
            out_vector[len(indices):] = self.char_vocab.mask_index
        else:
            out_vector = indices[:vector_length]
        
        return out_vector # 输出向量，和有表达力的长度（因为有些短句子后面全是pad）
    
    def vocab_size(self):
        return len(self.char_vocab)
    
    def decode_index(self,index_list):
        seq = []
        c = ''
        for s in index_list:
            seq.extend(self.char_vocab.lookup_index(int(s)))  
        return c.join(seq)

    @classmethod
    def from_list(cls,text_list):
        """Instantiate the vectorizer from the "total_text_list"        
        Args:
            total_text_list (list): a list of sentence,which is often contain 50 chinese charter
        Returns:
            an instance of the Vectorizer
        """
        char_vocab = SequenceVocabulary()

        for seq in text_list:
            for i, token in enumerate(jieba.cut(seq)):
                if i >= 40 :
                    break
                char_vocab.add_token(token)

        return cls(char_vocab)
    
    @classmethod
    def from_serializable(cls, contents):
        char_vocab = SequenceVocabulary.from_serializable(contents['char_vocab'])
        nat_vocab =  Vocabulary.from_serializable(contents['nationality_vocab'])

        return cls(char_vocab=char_vocab, nationality_vocab=nat_vocab)

    def to_serializable(self):
        return {'char_vocab': self.char_vocab.to_serializable(), 
                'nationality_vocab': self.nationality_vocab.to_serializable()}
```
